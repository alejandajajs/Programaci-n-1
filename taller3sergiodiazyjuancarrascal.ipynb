```python
class productos:
    def __init__(self, nombre, precio, categoria):
        self.nombre = nombre
        self.precio = precio
        self.categoria = categoria

    def __str__(self):
        return f"{self.nombre} - {self.precio} - {self.categoria}"

    def __repr__(self):
        return f"{self.nombre} - {self.precio} - {self.categoria}"

    def __eq__(self, other):
        if self.nombre == other.nombre:
            if self.categoria == other.categoria:
                return True
            else:
                return False

    def __lt__(self, other):
        if self.precio < other.precio:
            return True
        else:
            return False

    def __gt__(self, other):
        if self.precio > other.precio:
            return True
        else:
            return False


class Tienda:
    def __init__(self, tienda, lista_productos):
        self.tienda = tienda
        self.lista_productos = lista_productos

    def __str__(self):
        return f"{self.tienda} - {self.lista_productos}"

    def __repr__(self):
        return f"{self.tienda} - {self.lista_productos}"

    def agregar_producto(self, producto):
        self.lista_productos.append(producto)
        return self.lista_productos

    def eliminar_producto(self, nombre):
        for producto in self.lista_productos:
            if producto.nombre == nombre:
                self.lista_productos.remove(producto)
                return self.lista_productos
        return None

    def buscar_producto(self, nombre):
        for producto in self.lista_productos:
            if producto.nombre == nombre:
                return producto
        return None

    def producto_mas_caro(self):
        if not self.lista_productos:
            return None
        return max(self.lista_productos, key=lambda p: p.precio)

    def producto_mas_barato(self):
        if not self.lista_productos:
            return None
        return min(self.lista_productos, key=lambda p: p.precio)

    def mostrar_productos_por_categoria(self, categoria):
        return [p for p in self.lista_productos if p.categoria == categoria]

    def mostrar_productos_por_precio(self, precio_max):
        return [p for p in self.lista_productos if p.precio <= precio_max]


# Categorías
bebidas = "bebidas"
comidas = "comidas"
paquetes = "paquetes"

# Lista vacía por defecto
lista_productos = []

# EJEMPLO DE USO
# Crear productos
producto1 = productos("pastel de pollo", 2500, comidas)
producto2 = productos("jugo de naranja", 1500, bebidas)
producto3 = productos("combo almuerzo", 5000, paquetes)

# Crear tiendas
juan = Tienda("Tienda Juan", [producto1, producto2])
sergio = Tienda("Tienda Sergio", [producto3])

# Llamadas de prueba
print(juan.buscar_producto("pastel de pollo"))
print(sergio.buscar_producto("pastel de pollo"))
print(sergio.producto_mas_caro())
print(juan.producto_mas_caro())
print(sergio.producto_mas_barato())
print(juan.producto_mas_barato())
print(juan.mostrar_productos_por_categoria("comidas"))
print(sergio.mostrar_productos_por_categoria("comidas"))
print(sergio.mostrar_productos_por_precio(2500))
print(juan.mostrar_productos_por_precio(2500))
```
